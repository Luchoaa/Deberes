```python
class Queue:
    def __init__(self):
        self.s1 = []
        self.s2 = []
 
    def enQueue(self, x):
         
        while len(self.s1) != 0:
            self.s2.append(self.s1[-1])
            self.s1.pop()
            self.s1.append(x)
 
        while len(self.s2) != 0:
            self.s1.append(self.s2[-1])
            self.s2.pop()
 
    def deQueue(self):  
        
        if len(self.s1) == 0:
            print("Empty")
        else:       
            self.s1.pop()

 
    queue = Queue()
    queue.enQueue(1)
    queue.enQueue(2)
    queue.deQueue()
    
    print(queue.s1)
    print(queue.s2)

    
```

```python
from queue import Queue
 
class Stack:
     
    def __init__(self):
        self.q1 = Queue()
        self.q2 = Queue()
        
 
    def push(self, var):
        self.q2.put(var)
 
        while (not self.q1.empty()):
            self.q2.put(self.q1.queue[0])
            self.q1.get()
            
        self.qaux = self.q1
        self.q1 = self.q2
        self.q2 = self.qaux
 
    def pop(self):
        if (self.q1.empty()):
            return
        self.q1.get()

    def top(self):
        if (self.q1.empty()):
            return -1

    stack_1 = Stack()
    stack_1.push(8)
    stack_1.push(5) 
    stack_1.push(10)  
    stack_1.top()
    stack_1.pop()
```
